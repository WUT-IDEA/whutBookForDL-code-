```python
# create data
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(start=-10, stop=10, num=100000, dtype=np.float32)
```

## Linear
### $y=x$

```python
linear=x
plt.plot(x, linear)
plt.show()
```

## Sigmoid
### $y = \frac{1}{1+e^{-x}}$

```python
sigmoid=1/(1+np.exp(-x))
plt.plot(x,sigmoid)
plt.show()
```

## Relu
### $if\ x>0: y=x; else\ y=0$

```python
relu = [i if i >= 0 else 0 for i in x]
plt.plot(x, relu)
plt.show()
```

## Tanh
### $y=\frac{e^{x}-e^{-x}}{e^{x}+e^{-x}}$

```python
tanh = (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
plt.plot(x, tanh)
plt.show()
```
