```python
# -*- coding: utf-8 -*-

from __future__ import absolute_import
from __future__ import division
from __future__ import print_function
from six.moves import xrange
```

```python
import tensorflow as tf
import tensorflow.contrib.learn as learn
from tensorflow.contrib.learn.python.learn.estimators import model_fn as model_fn

from tensorflow.examples.tutorials.mnist import input_data


def fnn_model_fn(features, labels, mode):
    print('Input shape: %s, Ouput shape: %s.' % (features.shape, labels.shape))
    dense1 = tf.layers.dense(features, units=128, activation=tf.nn.relu)
    logits = tf.layers.dense(dense1, units=10)
    labels = tf.cast(labels, dtype=tf.int64)

    predictions = {"classes": tf.argmax(logits, axis=-1, name="classes_tensor"),
                   "probabilities": tf.nn.softmax(logits, name='softmax_tensor')}

    loss_op = tf.reduce_mean(tf.nn.sparse_softmax_cross_entropy_with_logits(labels=labels, logits=logits))

    if mode == learn.ModeKeys.TRAIN:
        train_op = tf.contrib.layers.optimize_loss(
            loss=loss_op,
            global_step=tf.contrib.framework.get_global_step(),
            learning_rate=1e-3,
            optimizer=tf.train.AdamOptimizer)
        return model_fn.ModelFnOps(mode=mode, predictions=predictions, loss=loss_op, train_op=train_op)
    else:
        return model_fn.ModelFnOps(mode=mode, predictions=predictions, loss=loss_op)

```

```python
def main(arg):
    # 建立一个learn.Estimator,模型保存或加载在model_dir中
    mnist = input_data.read_data_sets('../MNIST_data')
    classifier = learn.Estimator(model_fn=fnn_model_fn, model_dir="./model")
    tf.logging.set_verbosity(tf.logging.INFO) # 显示训练log

    # train
    # 将“classes_tensor”和“softmax_tensor”张量加入log中
    to_log = {"classes": "classes_tensor", "probabilities": "softmax_tensor"}
    log_hook = tf.train.LoggingTensorHook(to_log, every_n_iter=10)
    # 在原有模型的基础上，再进行10次训练
    classifier.fit(x=mnist.train.images, y=mnist.train.labels, batch_size=100, steps=10, monitors=[log_hook])

    # evaluate
    # 指定"accuracy"为测试指标
    metrics = {"accuracy": learn.MetricSpec(metric_fn=tf.metrics.accuracy, prediction_key="classes"), }
    eval_results = classifier.evaluate(x=mnist.test.images, y=mnist.test.labels, metrics=metrics)
    print(eval_results)


if __name__ == "__main__":
    tf.app.run()
```
